```python
#import enum to be able to create enumerators
from enum import Enum

#import pickle library to be able to serialize and deserialize files
import pickle


#create a class called BMCompany to store the companys information 
class BMCompany:
  # Initialize the class with companyName, and contactNumberr as attributes 
    def __init__(self, companyName, contactNumber):
        self._companyName = companyName
        self._contactNumber = contactNumber
        #create dictionaries to store the compants info about the employees, cars and sales
        self._employees = {}
        self._cars = {}
        self._sales = {}
        
        # Load dictionaries (employee dict, cars dict and sales dict) from binary files
        #use the try except block to factor in the scenario where the file is not found 
        try:
            with open("employees.pkl", "rb") as f:
                self._employees = pickle.load(f)
        except FileNotFoundError:
            print("employees.pkl file does not exist")

        try:
            with open("cars.pkl", "rb") as f:
                self._cars = pickle.load(f)
        except FileNotFoundError:
            print("cars.pkl file does not exist")

        try:
            with open("sales.pkl", "rb") as f:
                self._sales = pickle.load(f)
        except FileNotFoundError:
            print("sales.pkl file does not exist")

  # Setters and getters for the company name and contact number attributes

    def getCompanyName(self):
      return self._companyName
    
    def setCompanyName(self, companyName):
        self._companyName = companyName

    def getContactNumber(self):
      return self._contactNumber
    
    def setContactNumber(self, contactNumber):
        self._contactNumber = contactNumber

  #define function that allows the employee to be added and stored in the dictionary 
    def addEmployee(self, employee):
        # Add an employee to the dictionary of employees, using their employee ID as the key
        self._employees[employee.getEmployeeID()] = employee
        # Store the updated dictionary in binary files
        self.storeDictionaries()

  #define function that allows the car to be added and stored in the dictionary 
    def addCar(self, car):
      # Add an car to the dictionary of cars, using their car ID as the key
        self._cars[car.getCarID()] = car
         # Store the updated dictionary in binary files
        self.storeDictionaries()

    def addSale(self, sale):
      # Add the sale to the sales dictionary and use the sale number as the key in the sales dictionary
        self._sales[sale.getSaleNumber] = sale 
        # Store the updated _sales dictionary in a binary file using the storeDictionaries method    
        self.storeDictionaries()      

    def storeDictionaries(self):
        # Open "employees.pkl" binary file for writing and store the _employees dictionary using pickle.dump()
        with open("employees.pkl", "wb") as f:
            pickle.dump(self._employees, f)

        # Open "cars.pkl" binary file for writing and store the _cars dictionary using pickle.dump()
        with open("cars.pkl", "wb") as f:
            pickle.dump(self._cars, f)

        # Open "sales.pkl" binary file for writing and store the _sales dictionary using pickle.dump()
        with open("sales.pkl", "wb") as f:
            pickle.dump(self._sales, f)

# Define a Type class for enumeration with values as hatchback, sedan and suv
class Type(Enum):
    hatchback = "hatchback"
    sedan = "sedan"
    suv = "suv"

# Define a class called car
class Car:
  # Initialize the class with make, model, carID, price, carType, fuelCapacity, maxSpeed, color as attributes 
    def __init__(self, make, model, carID, price, carType, fuelCapacity, maxSpeed, color):
        self._make = make
        self._model = model
        self._carID = carID
        self._price = price
        self._carType = carType
        self._fuelCapacity = fuelCapacity
        self._maxSpeed = maxSpeed
        self._color = color
        
    # Setters and getters for the make, model, carID, price, carType, fuelCapacity, maxSpeed, color attributes 
    def getMake(self):
        return self._make
    
    def setMake(self, make):
        self._make = make
        
    def getModel(self):
        return self._model
    
    def setModel(self, model):
        self._model = model
        
    def getCarID(self):
        return self._carID
    
    def setCarID(self, carID):
        self._carID = carID
        
    def getPrice(self):
        return self._price
    
    def setPrice(self, price):
        self._price = price
        
    def getCarType(self):
        return self._carType
    
    def setCarType(self, carType):
        self._carType = carType

    def getFuelCapacity(self):
        return self._fuelCapacity
    
    def setFuelCapacity(self, fuelCapacity):
        self._fuelCapacity = fuelCapacity
        
    def getMaxSpeed(self):
        return self._maxSpeed
    
    def setMaxSpeed(self, maxSpeed):
        self._maxSpeed = maxSpeed
        
    def getColor(self):
        return self._color
    
    def setColor(self, color):
        self._color = color

  #create a __str__ function that wraps all the info of the class into a string 
    def __str__(self):
        return "Car ID: {}\nMake: {}\nModel: {}\nPrice: {}\nType: {}\nFuel Capacity: {}\nMax Speed: {}\nColor: {}".format(
            self._carID, self._make, self._model, self._price, self._carType.value, self._fuelCapacity, self._maxSpeed, self._color)
  
  #function that displays the car details when given a car id as an argument 
    def display_car_details(self, car_id):
      #it checks if the car if passed as an argument matches the attribute car id associated 
        if car_id == self._carID:
          #if so it prints the string with all the info about the car
            print(str(self))
        #else it gives a message that there is not car with that id
        else:
            print("Car with ID:", car_id, "is not found")

#define a class called sale
class Sale:
  # Initialize the class with saleNumber, employee, car, price as attributes 
    def __init__(self, saleNumber, employee, car, customer, price):
        self._saleNumber =saleNumber
        self._employee = employee
        self._car = car
        self._customer = customer
        self._price = price

        
    # Setters and getters for the saleNumber, employee, car, price as  attributes

    def getSaleNumber(self):
        return self._saleNumber
    
    def setSaleNumber(self, saleNumber):
        self._saleNumber = saleNumber

    def getEmployee(self):
        return self._employee
        
    def getCar(self):
        return self._car

    def setCustomer(self, customer):
        self._customer = customer 
    
    def getCustomer(self):
        return self._customer 


    def getPrice(self):
        return self._price
    
    def setPrice(self, price):
        self._price = price

#define an str method to combine the sales info into a string 
    def __str__(self):
      return f"Sale number: {self._saleNumber}\nEmployee: {self._employee}\nCar: {self._car}\nCustomer: {self._customer}\nPrice: {self._price}"


# Define a Gender class for enumeration with two values: male and female
class Gender(Enum):
  male = "male"
  female = "female"

# Define a Person class this is a parent class
class Person:
  # Initialize the class with a firstName, lastName, gender, dateOfBirth as attributes  
    def __init__(self, firstName, lastName, gender, dateOfBirth):
        self._firstName = firstName
        self._lastName = lastName
        self._gender = gender
        self._dateOfBirth = dateOfBirth
        
    # Setters and getters for the firstName, lastName, gender, dateOfBirth attributes  
    def getFirstName(self):
        return self._firstName
    
    def setFirstName(self, firstName):
        self._firstName = firstName
        
    def getLastName(self):
        return self._lastName
    
    def setLastName(self, lastName):
        self._lastName = lastName
        
    def getGender(self):
        return self._gender
    
    def setGender(self, gender):
        self._gender = gender
        
    def getDateOfBirth(self):
        return self._dateOfBirth
    
    def setDateOfBirth(self, dateOfBirth):
        self._dateOfBirth = dateOfBirth

# Define a class for a Customer, that is the child class of Person
class Customer(Person):
      # Initialize the class with a firstName, lastName, gender, dateOfBirth which are inherited as attributes 
    def __init__(self, firstName, lastName, gender, dateOfBirth, customerID):
        super().__init__(firstName, lastName, gender, dateOfBirth)
        self._customerID = customerID

    #define setters and getters
    def getCustomerID(self):
      return self._customerID
    
    def setCustomerID(self, customerID):
      self._customerID = customerID

    def __str__(self):
      return f'Customer ID: {self.getCustomerID()}\nName: {self.getFirstName()} {self.getLastName()}\nGender: {self.getGender()}\nDate of Birth: {self.getDateOfBirth()}'


# Define a JobTitle class for enumeration with two values: manager and salesperson
class JobTitle(Enum):
    manager = "Manager"
    salesperson = "Salesperson"

# Define a Department class for enumeration with these options 
class Department(Enum):
    Accounting = "Accounting"
    IT = "IT"
    Manufacturing = "Manufacturing"
    Human_Resources = "Human Resources"
    Sales = "Sales"
    Customer_Service = "Customer Service"
    Other = "Other"

# Define a class for a Employee, that is the child class of Person
class Employee(Person):
  # Initialize the class with a firstName, lastName, gender, dateOfBirth which are inherited as attributes 
  #and employeeID, basicSalary, department, jobTitle, passportNumber are the unique attributes of employee
    def __init__(self, firstName, lastName, gender, dateOfBirth, employeeID, basicSalary, department, jobTitle, passportNumber):
        super().__init__(firstName, lastName, gender, dateOfBirth)
        self._employeeID = employeeID
        self._basicSalary = basicSalary
        self._department = department
        self._jobTitle = jobTitle
        self._passportNumber = passportNumber
        #set manager as none 
        self._manager = None
        #define an empty list for salesperson to store the salesperons asssocaited for each manager 
        self._salespersons = []
        #define and empty list for sales to store the list of sales for each salesperson 
        self._sales = []

    #define setters and getters for employeeID, basicSalary, department, jobTitle, passportNumber attributes
    def getEmployeeID(self):
        return self._employeeID
    
    def setEmployeeID(self, employeeID):
        self._employeeID = employeeID
        
    def getBasicSalary(self):
        return self._basicSalary
    
    def setBasicSalary(self, basicSalary):
        self._basicSalary = basicSalary
    
    def getDepartment(self):
        return self._department
    
    def setDepartment(self, department):
        self._department = department
    
    def getJobTitle(self):
        return self._jobTitle
    
    def setJobTitle(self, jobTitle):
        self._jobTitle = jobTitle
    
    def getPassportNumber(self):
        return self._passportNumber
    
    def setPassportNumber(self, passportNumber):
        self._passportNumber = passportNumber
    
    #define a function that returns the list of sales
    def getSales(self):
        return self._sales

    #define a function that appends a sale to the list of sales
    def addSale(self, sale):
        self._sales.append(sale)

    #define a function that gets the manager of a salesperson 
    def get_manager(self):
        return self._manager

  #define a function that sets the manager of a salesperson 
    def set_manager(self, manager):
        self._manager = manager

        
    def calculate_salary(self):
        total_sales = sum(sale.getPrice() for sale in self._sales)
        salesperson_commission = total_sales * 0.065
        manager_commission = total_sales * 0.035
        company_profit = total_sales - salesperson_commission - manager_commission
        salary = self._basicSalary 

        if self._jobTitle == JobTitle.manager.value:
            salary += manager_commission
        elif self._jobTitle == JobTitle.salesperson.value:
            salary += salesperson_commission

        return salary


    #define a string function that wraps the information of the employee in a string 
    def __str__(self):
            return "Employee ID: {}\nName: {} {}\nGender: {}\nDate of Birth: {}\nBasic Salary: {}\nDepartment: {}\nJob Title: {}\nPassport Number: {}".format(
                self._employeeID, self._firstName, self._lastName, self._gender, self._dateOfBirth, self._basicSalary, self._department, self._jobTitle, self._passportNumber)
        

    #define a function to display a employee info with given an id 
    def display_employee_details(self, employee_id):
        # Check if the passed employee ID matches the attribute of the employee instance
        if employee_id == self._employeeID:
            # If the IDs match, print a string representation of the employee instance
            print(str(self))
        else:
            # If the IDs don't match, print a message saying that the employee was not found
            print("Employee with ID {} not found".format(employee_id))

    # Define a method to display sales details for a particular employee
    def display_sales_details(self, employee_id):
        # Check if the passed employee ID matches the attribute of the employee instance
        if employee_id == self._employeeID:
            # If the IDs match, check if the employee has any sales
            if len(self._sales) == 0:
                # If the employee has no sales, print a message saying so
                print("This employee has no sales.")
            else:
                # If the employee has sales, print a string with sales details for the employee
                print("Sales details for employee ID {} ({}, {}):".format(self._employeeID, self._firstName, self._lastName))
                # Loop through the employee's sales and print the car ID and sale price for each sale
                for sale in self._sales:
                    car_id = sale.getCar().getCarID()
                    sale_price = sale.getPrice()
                    print("Car ID: {}, Sale Price: {}".format(car_id, sale_price))
        else:
            # If the employee ID doesn't match the attribute, print a message saying the employee was not found
            print("Employee with ID {} not found".format(employee_id))


    #define a method to add a salesperson to a manager    
    def add_employee_to_manager(self, employee):
        # check if the employee is a Salesperson and the current employee is a Manager
        if self._jobTitle == JobTitle.manager and employee.getJobTitle() == JobTitle.salesperson:
            # set the manager of the employee to the current employee
            employee.set_manager(self)
            # add the employee to the salespersons list of the current employee
            self._salespersons.append(employee)
        else:
            # if the current employee is not a manager or the employee is not a salesperson, print an error message
            print("Manager can only manage salespersons.")
        
    def get_salespersons(self):
        # return the salespersons list of the current employee
        return self._salespersons

    def get_manager_name(self):
        if self._manager is None:
            # if the current employee has no manager, return this message message
            return "This employee has no manager."
        else:
            # return the full name of the manager of the current employee
            return "{} {}".format(self._manager.getFirstName(), self._manager.getLastName())
```
